```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
img = cv2.imread(r'C:\Users\kenneth\OneDrive\Desktop\PCV\PCV-LAB\Hisrogram_Equalization\image.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
kernel = np.ones((5,5), np.float32)/25
```

```python
mean = cv2.filter2D(gray_img, -1, kernel)
median = cv2.medianBlur(gray_img, 15)
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
```

```python
imgs = [gray_img, mean, median, laplacian]
title = ['Original Image','Mean Filter','Median Filter','Laplacian Filter']
```

```python
plt.figure(figsize = (10,10))
for i in range(len(imgs)):
    plt.subplot(2,2,i+1)
    plt.title(title[i])
    plt.imshow(imgs[i], cmap = 'gray')
plt.show()
```
