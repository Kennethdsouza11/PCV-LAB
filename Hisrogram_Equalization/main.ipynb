```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\kenneth\OneDrive\Desktop\PCV\PCV-LAB\Hisrogram_Equalization\image.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
```

```python
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
plt.title('Changed Image')
plt.imshow(img)
```

```python
gray_hist = cv2.calcHist(img_gray, [0], None, [256], [0,256])
plt.subplot(2,2,2)
plt.plot(gray_hist)
```

```python
plt.subplot(2,2,3)
plt.title('Histogram of Original Grascale image')
plt.hist(img_gray.ravel(), 256, [0,256])
```

```python
final_image = cv2.equalizeHist(img_gray)
plt.subplot(2,2,4)
plt.imshow(final_image, cmap = 'gray')
plt.show()
```
